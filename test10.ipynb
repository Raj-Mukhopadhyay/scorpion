```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import classification as cl
```

```python
data=pd.read_csv("classification_train.csv")
data.head()
```

```python
l=list()
for c in range(1,785):
    l.append(data[f"pixel{c}"])
xtemp=np.array(l)
x_train=np.transpose(xtemp)
y_train=np.array(data['label'])
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"pixel{i+1} \t{min(xtemp[i])}---{max(xtemp[i])}\n")
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
for i in range(len(xtemp)):
    if max(xtemp[i])>1:
        xtemp[i]=cl.scale_down(xtemp[i])
x_train=np.transpose(xtemp)
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"pixel{i+1} \t{min(xtemp[i])}---{max(xtemp[i])}\n")
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
xt=np.array(x_train)
```

```python
xtemp=np.transpose(xt)
for i in range(len(xtemp)):
    if max(xtemp[i])>1:
        xtemp[i]=cl.scale_down(xtemp[i])
xt=np.transpose(xtemp)
```

```python
cl.check_alpha_softmax(xt,y_train,0.001,2000)
```

```python
theta=cl.linear_softmax(xt,y_train,0.001,2000)
```

```python
domain=cl.Domain(y_train)
y_pre=cl.predict_softmax(xt,theta,domain)
```

```python
err=(y_pre-y_train)**2
error=(np.sum(err)/err.shape[0])**(1/2)
print(error)
```
