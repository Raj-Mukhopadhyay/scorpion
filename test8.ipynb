```python
import regression as rg
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data_draft.csv')
x=data['MinTemp']
y=data['MaxTemp']
n=len(x)
X=list()
Y=list()
for i in range(n):
    try:
        X.append(float(x[i]))
        Y.append(float(y[i]))
    except:
        continue
    
x_train=np.array(X)
y_train=np.array(Y)
```

```python
print(x_train)
print(y_train)
```

```python
w,b=rg.linear(x_train,y_train)
```

```python
print(x_train)
```

```python
y_pre=(w*x_train)+b
print("Prediction\tActual\n")
for i in range(y_pre.shape[0]):
    print(f"{y_pre[i]}\t{y_train[i]}\n")
```
