```python
import numpy as np
import pandas as pd
data=pd.read_csv("linear_train.csv")
data.head()
```

```python
l=list()
for c in range(65,85):
    l.append(data[chr(c)])
xtemp=np.array(l)
y_train=np.array(data['label'])
```

```python
import regression as rg
for i in range(len(xtemp)):
    if((max(xtemp[i])>=5)or(min(xtemp[i]<=-5))):
        xtemp[i]=rg.scale_down(xtemp[i])
x_train=np.transpose(xtemp)
```

```python
w,b=rg.linear_gen(x_train,y_train,4000)
```

```python
rg.set_alpha(x_train,y_train,4000)
```

```python
rg.gradient_movement(x_train,y_train,4000)
```
