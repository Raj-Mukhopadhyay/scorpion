```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# ------------------------------------------------TRAINING----------------------------------------------

```python
data=pd.read_csv('linear_train.csv')
data.head()
```

```python
l=list()
for c in range(65,85):
    l.append(data[chr(c)])
xtemp=np.array(l)
y_train=np.array(data['label'])
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"{chr(65+i)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
```

```python
import regression as rg
for i in range(len(xtemp)):
    if((max(xtemp[i])>=5)or(min(xtemp[i]<=-5))):
        xtemp[i]=rg.scale_down(xtemp[i])
x_train=np.transpose(xtemp)
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"{chr(65+i)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
```

```python
rg.check_alpha(x_train,y_train,0.01,600)
```

```python
w,b=rg.linear_gen(x_train,y_train,0.01,600)
```

```python
print(w)
print(b)
```

```python
x_train=np.transpose(xtemp)
l=[]
for i in range(len(x_train)):
    l.append(np.dot(w,x_train[i])+b)
y_pre=np.array(l)
for i in range(len(y_pre)):
    print(f"Prediction: {y_pre[i]};          Actual: {y_train[i]}\n")
```
