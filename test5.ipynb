```python
import numpy as np
import pandas as pd
data=pd.read_csv("linear_train.csv")
data.head()
```

```python
l=list()
for c in range(65,85):
    l.append(data[chr(c)])
x_train=np.transpose(np.array(l))
y_train=np.array(data['label'])
```

```python
import regression as rg
w_final,b_final=rg.linear_gen(x_train,y_train)
print("+-------------------------------------------------+")
m=len(x_train)
for i in range(m):
    print(f"prediction: {np.dot(x_train[i], w_final) + b_final}, target value: {y_train[i]}")
print("+-------------------------------------------------+")
```
