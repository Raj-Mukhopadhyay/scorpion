```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import regression as rg
```

# -----------------------------------------------TRAINING-----------------------------------------------

```python
data=pd.read_csv("polynomial_train.csv")
data.head()
```

```python
l=list()
for c in range(65,68):
    l.append(data[chr(c)])
xtemp=np.array(l)
x_train=np.transpose(xtemp)
y_train=np.array(data['label'])
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"{chr(65+i)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
import regression as rg
for i in range(len(xtemp)):
    if((max(xtemp[i])>=5)or(min(xtemp[i]<=-5))):
        xtemp[i]=rg.scale_down(xtemp[i])
if((max(y_train)>=5)or(min(y_train)<=-5)):
    y_train=rg.scale_down(y_train)
x_train=np.transpose(xtemp)
```

```python
print("Column\tRange\n")
for i in range(len(xtemp)):
    print(f"{chr(65+i)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
print(x_train.shape)
x_train_trans=rg.transform(x_train,[5,4,4])
print(x_train_trans.shape)
```

```python
xtemp=np.transpose(x_train_trans)
order=[5,4,4]
c=0
c1=0
print("Column\tRange\n")
for i in range(len(xtemp)):
    if c1==0:
        print(f"{chr(65+c)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
    else:
        print(f"{chr(65+c)}^{c1+1}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
    if c1==order[c]-1:
        c1=-1
        c+=1
        
    c1+=1
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
for i in range(len(xtemp)):
    if((max(xtemp[i])>=5)or(min(xtemp[i]<=-5))):
        xtemp[i]=rg.scale_down(xtemp[i])
if((max(y_train)>=5)or(min(y_train)<=-5)):
    y_train=rg.scale_down(y_train)
x_train_trans=np.transpose(xtemp)
```

```python
order=[5,4,4]
c=0
c1=0
print("Column\tRange\n")
for i in range(len(xtemp)):
    if c1==0:
        print(f"{chr(65+c)}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
    else:
        print(f"{chr(65+c)}^{c1+1}\t{min(xtemp[i])}---{max(xtemp[i])}\n")
    if c1==order[c]-1:
        c1=-1
        c+=1
        
    c1+=1
print("_____________________________________________________________")
print(f"label\t{min(y_train)}---{max(y_train)}\n")
```

```python
rg.check_alpha(x_train_trans,y_train,0.05,150)
```

```python
w,b=rg.linear_gen(x_train_trans,y_train,0.05,150)
```

```python
w1,b1=rg.polynomial(x_train,y_train,[5,4,4],0.05,150)
```

```python
l=list()
for x in x_train_trans:
    l.append(np.dot(w,x)+b)
y_pre=np.array(l)
```
